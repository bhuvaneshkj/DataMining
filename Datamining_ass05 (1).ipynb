```python
import pandas as pd
import numpy as np
from sklearn.model_selection  import train_test_split
```

```python
df=pd.read_csv(r'C:\Users\chat2\Downloads\ASS05_Data.csv')
```

```python
df.head()
df1=df.rename(columns={'SalePrice':'TrueTarget'})
df1
```

```python
X=df1.loc[:, df1.columns != 'TrueTarget']
Y=df1.loc[:,'TrueTarget']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,accuracy_score,mean_squared_error
import statsmodels.api as sm
score_train=[]
mse_train=[]
r_sqaured=[]
results1=[]
score_test=[]
mse_validation=[]
for i in range(0,5):
    
    xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.30)
    
    reg = LinearRegression().fit(xtrain, ytrain)
    score_train.append(reg.score(xtrain,ytrain))
    mse_train.append(mean_squared_error(ytrain,reg.predict(xtrain)))
    model = sm.OLS(ytrain,xtrain)
    results = model.fit()
    results1.append(results)
    score_test.append(reg.score(xtest,ytest))
    r_sqaured.append(results.rsquared)
    mse_validation.append(mean_squared_error(ytest,reg.predict(xtest)))

for i in range(0,len(results1)):
    print(results1[i].summary())

```

```python
print("Rsqaured of Training Set = "+str(r_sqaured))
print("mean squared error of Training Set = "+str(mse_train))
print("Training error score = "+str(score_train))
```

```python
print("mean squared error of Test Set = "+str(mse_validation))
print("Testing error score = "+str(score_test))
```
